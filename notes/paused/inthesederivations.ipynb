# In these derivations

Row vectors are denoted as bold lower-case Latin or Greek, e.g. "$\mathbf a$". Column vectors are denoted as tranposed row vectors, e.g. "$\mathbf a^\top$". Matrices and linear operators are denoted as bold, upper-case Latin and Greek e.g. "$\mathbf A$". The $N\times N$ identity is denoted as $\mathbf I_n$. The $N\times K$ matrix of 1s is denoted as $\mathbb 1_{nk}$, or simply $\mathbb 1$ when dimensions are unambiguous. The vector of all 1s is denoted similarly. Element-wise multiplication is denoted with $a \circ b$. Element-wise division is denoted as $\frac a b$.

We use the following identities: 

The trace is invariant under cyclic permutation and transpose:

$$
\operatorname{tr}[\mathbf A\mathbf B^\top] =
\operatorname{tr}[\mathbf B^\top \mathbf A] =
\operatorname{tr}[\mathbf A^\top \mathbf B] =
\operatorname{tr}[\mathbf B \mathbf A^\top]
$$

The trace of the product of a matrix and a diagonal matrix is

$$
\operatorname{tr}[\mathbf A\operatorname{diag}(\mathbf d)] = \operatorname{diag}[\mathbf A]^\top \mathbf d
$$

The product of a diagonal matrix and a vector can be computed elementwise

$$
\operatorname{diag}[\mathbf d]\,\mathbf v = \mathbf d \circ \mathbf v
$$

If $\mathbf D = \operatorname{diag}[\mathbf d]$ is diagonal, then 

$$
\operatorname{diag}[\mathbf A^\top \mathbf D \mathbf B] = (\mathbf A\circ\mathbf B)^\top \mathbf d
$$

This can be verified by considering the matrix product element-wise:

$$
(\mathbf A^\top\mathbf D \mathbf B)_{ii} 
= \sum_{kl} \mathbf A_{ki} \mathbf D_{kl} \mathbf B_{li}
= \sum_{k} \mathbf A_{ki} \mathbf B_{ki} \mathbf d_{k} = (\mathbf A\circ\mathbf B)^\top \mathbf d
$$

Let $\mathbf D = \operatorname{diag}[\mathbf d]$ be a diagonal matrix, then: 

To multiply each row of $\mathbf A$ by the elements of a vector:
$$
(\mathbf D \mathbf A)_{ij} = \sum_k \mathbf D_{ik} \mathbf A_{kj} = \mathbf d_i \mathbf A_{ij}
$$

To multiply each column of $\mathbf A$ by the elements of a vector:
$$
(\mathbf A \mathbf D)_{ij} = \sum_k \mathbf A_{ik} \mathbf D_{kj} = \mathbf d_j \mathbf A_{ij}
$$

To sum columns within each row of a matrix: 

$$
(\mathbf M\mathbb 1)_{i} = \sum_{k} \mathbf M_{ik} \mathbb 1_{k} = \sum_k \mathbf M_{ik}
$$

To sum rows within each column of a matrix: 

$$
(\mathbb 1^\top \mathbf M)_{j} = \sum_{k} \mathbb 1_{k} \mathbf M_{kj} = \sum_k \mathbf M_{kj}
$$

For matrices, the following denote the inner / scalar / Frombenius product:
    
$$
\left<\mathbf A,\mathbf B\right>
=
\left<\mathbf A|\mathbf B\right>
= 
\operatorname{tr}\left[\mathbf A^\top\mathbf B\right]
=
\mathbf A:\mathbf B
= \textstyle\sum_{ij}\mathbf A_{ij}\mathbf B_{ij}
= \mathbf A_{ij}\mathbf B_{ij}
$$

Goulob and Pereya (1972) Eq. 4.12 gives the derivative of a fixed-rank pseudoinverse: 

$$
\partial_x \mathbf A^+(x) =
- \mathbf A^+ (\partial_x \mathbf A) \mathbf A^+
+ \mathbf A^+ \mathbf A{^+}^\top (\partial_x \mathbf A)^\top (1-\mathbf A \mathbf A^+)
+ (1-\mathbf A^+ \mathbf A)(\partial_x \mathbf A)^\top \mathbf A{^+}^\top \mathbf A^+
$$

Assuming $\mathbf A$ is a tall, thin matrix $N\times K$, $K<N$, then $\mathbf A\mathbf A^+$ is shape $N\times N$ and $\mathbf A^+ \mathbf A$ is shape $K\times K$. In this construction $\mathbf A^+\mathbf A = \mathbf I_k$ and the final term vanishes. The derivative of the pseudoinverse can now be written as: 

\begin{equation}\begin{aligned}
\partial_x \mathbf A^+(x) 
&=
- \mathbf A^+ (\partial_x \mathbf A) \mathbf A^+
+ \mathbf A^+ \mathbf A{^+}^\top (\partial_x \mathbf A)^\top ( \mathbf I_n - \mathbf A\mathbf A^+ )
\end{aligned}\end{equation}

In this form the relationship to the derivative of a matrix inverse is clear. The derivative is similar, with the pseudoinverse $\mathbf A^+$ replacing the true inverse $\mathbf A^{-1}$. There is also an additional term 
$\mathbf A^+ \mathbf A{^+}^\top (\partial_x \mathbf A)^\top \mathbf P$ that is related to the null-space of $\mathbf A$.
