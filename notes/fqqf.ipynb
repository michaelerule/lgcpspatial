Neglecting terms that do not depend on $\mathbf X$, the loss to be minimized is:
\begin{equation}\begin{aligned}
\mathcal L(\mathbf Q)
&
=
\mathbf n^\top \exp\left(\boldsymbol\mu + \tfrac 1 2 \operatorname{diag}\left[\mathbf F^\top \mathbf Q\mathbf Q^\top \mathbf F\right]\right)
+
\tfrac 1 2 
\left\{
\operatorname{tr}\left[
\mathbf F \boldsymbol\Sigma_0^{-1}\mathbf F^\top \mathbf Q\mathbf Q^\top 
\right] 
- 
\ln|\mathbf Q\mathbf Q^\top|\right\}
\end{aligned}\end{equation}

\begin{equation}\begin{aligned}
{\partial_{\mathbf Q_{ij}}}\operatorname{tr}\left[ \mathbf Q^\top \mathbf F \mathbf D \mathbf F^\top \mathbf M \right]
&=
{\partial_{\mathbf Q_{ij}}}\left[\mathbf Q^\top \mathbf F \mathbf D \mathbf F^\top \mathbf M\right]kk
\\&=
{\partial_{\mathbf Q_{ij}}}(\mathbf Q^\top \mathbf F)_{kl} \mathbf D_l (\mathbf F^\top \mathbf M)_{lk}
\\&=
\mathbf F \mathbf D \mathbf F^\top \mathbf M + (\mathbf Q^\top \mathbf F)_{kl} ({\partial_{\mathbf Q_{ij}}}\mathbf D_l) (\mathbf F^\top \mathbf M)_{lk}
\\&=
\mathbf F \mathbf D \mathbf F^\top \mathbf M + (\mathbf Q^\top \mathbf F)_{kl} \left[{\tfrac12}\mathbf D_l {\partial_{\mathbf Q_{ij}}}(\mathbf F^\top \mathbf Q \mathbf Q^\top \mathbf F)_{ll} \right] (\mathbf F^\top \mathbf M)_{lk}
\\&=
\mathbf F \mathbf D \mathbf F^\top \mathbf M + (\mathbf Q^\top \mathbf F)_{kl} \left[{\tfrac12}\mathbf D_l {\partial_{\mathbf Q_{ij}}}(\mathbf F^\top_{lm} \mathbf Q_{mn} \mathbf Q^\top_{no} \mathbf F_{ol}) \right] (\mathbf F^\top \mathbf M)_{lk}
\\&=
\mathbf F \mathbf D \mathbf F^\top \mathbf M + (\mathbf Q^\top \mathbf F)_{kl} {{\tfrac12}\mathbf D_l \left[\mathbf F^\top_{lm} (\left[i,j=o,n\right] \mathbf Q_{mn} + \left[ij{=}mn\right] \mathbf Q_{on}) \mathbf F_{ol}\right] } (\mathbf F^\top \mathbf M)_{lk}
\\&=
\mathbf F \mathbf D \mathbf F^\top \mathbf M + (\mathbf Q^\top \mathbf F)_{kl} {{\tfrac12}\mathbf D_l \left[\left[ij{=}on\right] \mathbf F^\top_{lm} \mathbf Q_{mn} \mathbf F_{ol} + \left[ij{=}mn\right] \mathbf F^\top_{lm} \mathbf Q_{on} \mathbf F_{ol}\right] } (\mathbf F^\top \mathbf M)_{lk}
\\&=
\mathbf F \mathbf D \mathbf F^\top \mathbf M + (\mathbf Q^\top \mathbf F)_{kl} {{\tfrac12}\mathbf D_l \left[\mathbf F^\top_{lm} \mathbf Q_{mj} \mathbf F_{il} + \mathbf F^\top_{li} \mathbf Q_{oj} \mathbf F_{ol}\right] } (\mathbf F^\top \mathbf M)_{lk}
\\&=

\mathbf F \mathbf D \mathbf F^\top \mathbf M + (\mathbf Q^\top \mathbf F)_{kl} \mathbf D_l (\mathbf F^\top \mathbf Q)_{lj} \mathbf F_{il} (\mathbf F^\top \mathbf M)_{lk}

\mathbf F \mathbf D \mathbf F^\top \mathbf M + \mathbf F_{il} \mathbf D_l (\mathbf F^\top \mathbf M\mathbf Q^\top \mathbf F)_{ll} (\mathbf F^\top \mathbf Q)_{lj} 
\\&=
\mathbf F \mathbf D \mathbf F^\top \mathbf M + \mathbf F \mathbf D \operatorname{diag}(\mathbf F^\top \mathbf M\mathbf Q^\top \mathbf F) \mathbf F^\top \mathbf Q
\end{aligned}\end{equation}
